```python
from triaxial_ellipsoid import ellipsoid_surface
from triaxial_ellipsoid import plot_ellipsoid
from triaxial_ellipsoid import plot_ellipsoid_planview

%matplotlib inline
import matplotlib_inline
matplotlib_inline.backend_inline.set_matplotlib_formats('retina')
```

```python
#Â Source parameters
x_centre = 1000
y_centre  = -1500
z_centre  = -4000
x_radius  = 1000
y_radius  = 1500
z_radius  = 500
phi = 60 # azimuth angle from North, about Z-axis
theta = 45 # inclination angle from horizontal, about Y-axis
psi = -30 # second rotation about Z-axis
```

```python
# Coords @ Rz(0) @ Ry(0) @ Rz(phi)
# -> rotate around Z-axis (azimuthally) by phi
X, Y, Z = ellipsoid_surface(x_centre, y_centre, z_centre, x_radius, y_radius, z_radius, phi, 0, 0)

plot_ellipsoid(X, Y, Z)
plot_ellipsoid_planview(X, Y, Z)
```

```python
# Coords @ Rz(0) @ Ry(theta) @ Rz(phi)
# -> rotate around Z-axis (azimuthally) by phi, then rotate around Y by theta
X, Y, Z = ellipsoid_surface(x_centre, y_centre, z_centre, x_radius, y_radius, z_radius, phi, theta, 0)

plot_ellipsoid(X, Y, Z)
plot_ellipsoid_planview(X, Y, Z)
```

```python
# Coords @ Rz(psi) @ Ry(theta) @ Rz(phi)
# -> rotate around Z-axis (azimuthally) by phi, then rotate around Y by theta, then rotate around ellipse Z-axis by psi
X, Y, Z = ellipsoid_surface(x_centre, y_centre, z_centre, x_radius, y_radius, z_radius, phi, theta, psi)

plot_ellipsoid(X, Y, Z)
plot_ellipsoid_planview(X, Y, Z)
```
